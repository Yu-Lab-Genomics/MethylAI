```python
import pandas as pd
import numpy as np
import timeit

# 创建一个大 DataFrame
df = pd.DataFrame(np.random.randn(100_000, 10))
df.index = [f"row_{i}" for i in range(100_000)]  # 使用字符串索引，放大 loc 的查找开销
select_value = df.iloc[50000, 0]

# 测试 iloc
def test_iloc():
    index = np.where(df[0] == select_value)[0].tolist()
    return df.iloc[index, [2,3,4]]

# 测试 loc
def test_loc():
    index = df[df[0] == select_value].index
    return df.loc[index, [2,3,4]]

a = test_iloc()
b = test_loc()
print(a)
print(b)
```

```python
# 计时
time_iloc = timeit.timeit(test_iloc, number=1000)
time_loc = timeit.timeit(test_loc, number=1000)

print(f"iloc 平均时间: {time_iloc:.6f} 秒")
print(f"loc 平均时间: {time_loc:.6f} 秒")
```
